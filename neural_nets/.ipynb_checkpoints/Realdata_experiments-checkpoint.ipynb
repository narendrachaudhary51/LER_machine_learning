```python
#path = '/home/grads/n/narendra5/Desktop/Programs/LER_machine_learning/'
path = r"C:\\Users\\naren\\Documents\\LER\\"

#model = load_model(path + 'models/' + 'EDGEnet_run_epoch_4.h5')

model = load_model(path + 'LER_machine_learning\\models\\' + 'LineNet_image4_round_L2_rotation_epoch_5.h5')
#model.summary()
```

```python
noisy_file = r'C:\Users\naren\Documents\LER\For_Texas_AM_University\25_100NM_CH.jpg'
# noisy_file = r'C:\Users\naren\Documents\LER\For_Texas_AM_University\data4' + \
#             '\FhG_IPMS_CDSEM_LER_Jan2020\E2OU920400_16_E0142_002_LER_LITHO_1\PatternFOV' + \
#             '\E0142_002_LER_E0142_002_LER_UP_FX0_FY0_RR1_TR1_Ver1.jpg'

imnoisy_color = np.array(Image.open(noisy_file))
#imnoisy = color.rgb2gray(imnoisy_color)
imnoisy = cv2.cvtColor(imnoisy_color, cv2.COLOR_BGR2GRAY)

#imnoisy = np.array(Image.open(noisy_file))

FOV = 1

# plt.figure(figsize = (8,8))
# plt.imshow(imnoisy[444:,:], cmap = 'gray')
# plt.show()



# imnoisy = cv2.resize(imnoisy, dsize=(120,111))                 #FOV = 0.5
# imnoisy = np.pad(imnoisy, ((0, 145), (0, 136)), 'constant')      #FOV = 0.5

# imnoisy = (imnoisy[110:366,184:440])                    #FOV = 2

imnoisy = imnoisy / 256

print(imnoisy.shape)

print(imnoisy.min(), imnoisy.max())

if FOV == 1:
    imnoisy = imnoisy[:444,:]
    #imnoisy = cv2.resize(imnoisy, dsize=(240,222))                 #FOV = 1
    imnoisy = imnoisy[::2,::2]
    imnoisy = np.pad(imnoisy, ((0, 34), (0, 16)), 'constant')      #FOV = 1
    
    #imnoisy = cv2.resize(imnoisy, dsize=(240,240))                    #FOV = 1 and no cutting
    #imnoisy = np.pad(imnoisy, ((0, 16), (0, 16)), 'constant')        #FOV = 1 and no cutting
    
    imnoisy = imnoisy.reshape(1,256,256,1)
    predict = np.zeros([1,256,256,2])
    predict = model.predict(imnoisy)
    predict = predict.reshape(256,256,2)
    imnoisy = imnoisy.reshape(256,256)
    
if FOV == 2:
    #imnoisy = imnoisy[:444,:]
    #imnoisy = np.pad(imnoisy, ((0, 36), (0, 0)), 'constant') 
    
    imnoisy = imnoisy.reshape(1,480,480,1)
    predict = np.zeros([1,480,480,2])
    patch = np.zeros([4,256,256,2])
    patch[0,:,:] = model.predict(imnoisy[:,0:256,0:256,:])
    patch[1,:,:] = model.predict(imnoisy[:,224:,0:256,:])
    patch[2,:,:] = model.predict(imnoisy[:,0:256,224:,:])
    patch[3,:,:] = model.predict(imnoisy[:,224:,224:,:])
    
    mid_x = 230
    mid_y = 230
    predict[:,0:mid_x, 0:mid_y,:] = patch[0,0:mid_x, 0:mid_y,:]
    predict[:,mid_x:, 0:mid_y,:] = patch[1,(mid_x - 224):, 0:mid_y,:]
    predict[:,0:mid_x, mid_y:,:] = patch[2,0:mid_x, (mid_y - 224):,:]
    predict[:,mid_x:, mid_y:,:] = patch[3,(mid_x - 224):, (mid_y - 224):,:]
    predict = predict.reshape(480,480,2)
    imnoisy = imnoisy.reshape(480,480)

print(predict.shape)
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16, 16), sharex=True,
                       sharey=True)


ax[0].imshow(imnoisy, cmap = 'gray')
ax[0].set_title('(a) Real SEM Image', fontsize=18)
ax[1].imshow(predict[:,:,0], cmap = 'gray')
ax[1].set_title('(b) Predicted Denoised Image', fontsize=18)
ax[2].imshow(predict[:,:,1], cmap = 'gray')
ax[2].set_title('(c) Predicted Edge Image', fontsize=18)
plt.show()
```

```python
from scipy.signal import find_peaks    # To get maxima
nedges = 6
length = 400
predicted_edges = np.zeros((nedges, length))
CD = np.zeros((nedges//2, length))
dist = 15       # initial minimum distance between the peaks
over = 0          # flag peak finishing peak search


for i in range(length):
    peaks, _ = find_peaks(predict[i,0:,0], height = 0.40, distance = dist)
    predicted_edges[:,i] = peaks[:nedges]

#print(predicted_edges[3,:].shape)
plt.figure(figsize = (14,8))
plt.plot(predicted_edges[:].T)
plt.show()

pixel_size = (2000/480)
#pixel_size = (1000/480)

for i in range(nedges//2):
    CD[i,:] = predicted_edges[i+1,:] - predicted_edges[i,:]

print(predicted_edges.shape, CD.shape)

print((3*predicted_edges.std(axis = 1)*pixel_size*np.sqrt(length/(length-2))))
print("Mean LER: ", (3*predicted_edges.std(axis = 1)*pixel_size*np.sqrt(length/(length-2))).mean())

print((3*CD.std(axis = 1)*pixel_size*np.sqrt(length/(length-1))))
print("Mean LWR: ", (3*CD.std(axis = 1)*pixel_size*np.sqrt(length/(length-1))).mean())

#means = predicted_edges.mean(axis =1)
#for i in range(nedges):
#    predicted_edges[i,:] = predicted_edges[i,:] - means[i]
#print(predicted_edges.flatten().std()*(2000/480))
```
