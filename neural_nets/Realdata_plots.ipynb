```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel(r'C:\Users\naren\Documents\LER\For_Texas_AM_University\data4\FhG_IPMS_CDSEM_LER_Jan2020\Neural_networks_LER_experiment (1).xls', index_col=0) 
print(df)
```

```python
df_iso = df.loc[df['MeasName'] == 'iso']
df_dense = df.loc[df['MeasName'] == 'dense']
```

```python
LER_Bottom_up_iso = np.zeros((2,8))
LER_denoised_up_iso = np.zeros((2,8))
LER_edge_up_iso = np.zeros((2,8))

LER_Bottom_right_iso = np.zeros((2,8))
LER_denoised_right_iso = np.zeros((2,8))
LER_edge_right_iso = np.zeros((2,8))

for wafer in range(16,24):
    mean_series = df_iso.loc[(df_iso['Wafer'] == wafer) & (df_iso['Target_Name'] == 'E0142_002_LER_UP')].mean()
    std_series = df_iso.loc[(df_iso['Wafer'] == wafer) & (df_iso['Target_Name'] == 'E0142_002_LER_UP')].std()
    LER_Bottom_up_iso[0, wafer - 16] = mean_series['LER_Bottom'] 
    LER_Bottom_up_iso[1, wafer - 16] = std_series['LER_Bottom']
    
    LER_denoised_up_iso[0, wafer - 16] = mean_series['LER_denoised'] 
    LER_denoised_up_iso[1, wafer - 16] = std_series['LER_denoised']
    
    LER_edge_up_iso[0, wafer - 16] = mean_series['LER_edge'] 
    LER_edge_up_iso[1, wafer - 16] = std_series['LER_edge']
    
#print(LER_denoised_up_iso)

for wafer in range(16,24):
    mean_series = df_iso.loc[(df_iso['Wafer'] == wafer) & (df_iso['Target_Name'] == 'E0142_002_LER_RIGHT')].mean()
    std_series = df_iso.loc[(df_iso['Wafer'] == wafer) & (df_iso['Target_Name'] == 'E0142_002_LER_RIGHT')].std()
    LER_Bottom_right_iso[0, wafer - 16] = mean_series['LER_Bottom'] 
    LER_Bottom_right_iso[1, wafer - 16] = std_series['LER_Bottom']
    
    LER_denoised_right_iso[0, wafer - 16] = mean_series['LER_denoised'] 
    LER_denoised_right_iso[1, wafer - 16] = std_series['LER_denoised']
    
    LER_edge_right_iso[0, wafer - 16] = mean_series['LER_edge'] 
    LER_edge_right_iso[1, wafer - 16] = std_series['LER_edge']
    
#print(LER_denoised_right_iso)
```

```python
LER_Bottom_up_dense = np.zeros((2,8))
LER_denoised_up_dense = np.zeros((2,8))
LER_edge_up_dense = np.zeros((2,8))

LER_Bottom_right_dense = np.zeros((2,8))
LER_denoised_right_dense = np.zeros((2,8))
LER_edge_right_dense = np.zeros((2,8))

for wafer in range(16,24):
    mean_series = df_dense.loc[(df_dense['Wafer'] == wafer) & (df_dense['Target_Name'] == 'E0142_002_LER_UP')].mean()
    std_series = df_dense.loc[(df_dense['Wafer'] == wafer) & (df_dense['Target_Name'] == 'E0142_002_LER_UP')].std()
    LER_Bottom_up_dense[0, wafer - 16] = mean_series['LER_Bottom'] 
    LER_Bottom_up_dense[1, wafer - 16] = std_series['LER_Bottom']
    
    LER_denoised_up_dense[0, wafer - 16] = mean_series['LER_denoised'] 
    LER_denoised_up_dense[1, wafer - 16] = std_series['LER_denoised']
    
    LER_edge_up_dense[0, wafer - 16] = mean_series['LER_edge'] 
    LER_edge_up_dense[1, wafer - 16] = std_series['LER_edge']
    
#print(LER_edge_up_iso)

for wafer in range(16,24):
    mean_series = df_dense.loc[(df_dense['Wafer'] == wafer) & (df_dense['Target_Name'] == 'E0142_002_LER_RIGHT')].mean()
    std_series = df_dense.loc[(df_dense['Wafer'] == wafer) & (df_dense['Target_Name'] == 'E0142_002_LER_RIGHT')].std()
    LER_Bottom_right_dense[0, wafer - 16] = mean_series['LER_Bottom'] 
    LER_Bottom_right_dense[1, wafer - 16] = std_series['LER_Bottom']
    
    LER_denoised_right_dense[0, wafer - 16] = mean_series['LER_denoised'] 
    LER_denoised_right_dense[1, wafer - 16] = std_series['LER_denoised']
    
    LER_edge_right_dense[0, wafer - 16] = mean_series['LER_edge'] 
    LER_edge_right_dense[1, wafer - 16] = std_series['LER_edge']
    
#print(LER_edge_right_iso)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(16, 10), sharex=True,
                       sharey=False)
ax[0,0].plot(range(16,24), LER_Bottom_up_iso[0], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[0,0].plot(range(16,24), LER_Bottom_up_dense[0], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[0,0].plot(range(16,24), LER_Bottom_right_iso[0], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[0,0].plot(range(16,24), LER_Bottom_right_dense[0], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[0,0].set_title('(a) LER Bottom Mean', fontsize=16)
ax[0,0].set_ylabel('LER (nm)', fontsize=12)
ax[0,0].set_xlabel('Wafer Number', fontsize=12)
ax[0,0].set_ylim(4,14)
ax[0,0].legend()

ax[0,1].plot(range(16,24), LER_denoised_up_iso[0], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[0,1].plot(range(16,24), LER_denoised_up_dense[0], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[0,1].plot(range(16,24), LER_denoised_right_iso[0], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[0,1].plot(range(16,24), LER_denoised_right_dense[0], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[0,1].set_title('(b) LER Denoised Mean (NN)', fontsize=16)
ax[0,1].set_ylabel('LER (nm)', fontsize=12)
ax[0,1].set_xlabel('Wafer Number', fontsize=12)
ax[0,1].set_ylim(4,14)
ax[0,1].legend()

ax[0,2].plot(range(16,24), LER_edge_up_iso[0], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[0,2].plot(range(16,24), LER_edge_up_dense[0], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[0,2].plot(range(16,24), LER_edge_right_iso[0], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[0,2].plot(range(16,24), LER_edge_right_dense[0], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[0,2].set_title('(c) LER Edge Mean (NN)', fontsize=16)
ax[0,2].set_ylabel('LER (nm)', fontsize=12)
ax[0,2].set_xlabel('Wafer Number', fontsize=12)
ax[0,2].set_ylim(4,14)
ax[0,2].legend()

ax[1,0].plot(range(16,24), LER_Bottom_up_iso[1], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[1,0].plot(range(16,24), LER_Bottom_up_dense[1], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[1,0].plot(range(16,24), LER_Bottom_right_iso[1], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[1,0].plot(range(16,24), LER_Bottom_right_dense[1], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[1,0].set_title('(d) LER Bottom Dev.', fontsize=16)
ax[1,0].set_ylabel('LER (nm)', fontsize=12)
ax[1,0].set_xlabel('Wafer Number', fontsize=12)
ax[1,0].set_ylim(0,6)
ax[1,0].legend()

ax[1,1].plot(range(16,24), LER_denoised_up_iso[1], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[1,1].plot(range(16,24), LER_denoised_up_dense[1], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[1,1].plot(range(16,24), LER_denoised_right_iso[1], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[1,1].plot(range(16,24), LER_denoised_right_dense[1], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[1,1].set_title('(e) LER Denoised Dev. (NN)', fontsize=16)
ax[1,1].set_ylabel('LER (nm)', fontsize=12)
ax[1,1].set_xlabel('Wafer Number', fontsize=12)
ax[1,1].set_ylim(0,6)
ax[1,1].legend()

ax[1,2].plot(range(16,24), LER_edge_up_iso[1], 'o', markersize=10, color='tab:blue', fillstyle='none', label = 'up iso')
ax[1,2].plot(range(16,24), LER_edge_up_dense[1], 'o', markersize=6, color='tab:blue', label = 'up dense')
ax[1,2].plot(range(16,24), LER_edge_right_iso[1], 'o', markersize=10, color='tab:red', fillstyle='none', label = 'right iso')
ax[1,2].plot(range(16,24), LER_edge_right_dense[1], 'o', markersize=6, color='tab:red', label = 'right dense')
ax[1,2].set_title('(f) LER Edge Dev. (NN)', fontsize=16)
ax[1,2].set_ylabel('LER (nm)', fontsize=12)
ax[1,2].set_xlabel('Wafer Number', fontsize=12)
ax[1,2].set_ylim(0,6)
ax[1,2].legend()

plt.show()
```
